### Let's Work on a Movie Dataset
-Make a scatter plot between Length of Movie Title(X-Axis) and Frequency of occurence of that (along Y-axis)

```python
df = pd.read_csv('movie_metadata.csv.zip')
```

```python
df
```

```python
columns = df.columns
```

```python
columns
```

```python
titles = list(df.get('movie_title'))
print(titles)
```

```python
titles = list(df.get('imdb_score'))
print(titles)
```

```python
titles = df.get("imdb_score")
print(titles)
```

```python
df.shape
```

```python
titles = df.get('movie_title')
```

```python
for t in titles:
    length = len(t)
    print(length)
```

```python
freq_title = {}

for t in titles:
    length = len(t)

    try:
        freq_title[length] += 1
    except:
        freq_title[length] = 1

print(freq_title)            
```

```python
x = np.array(list(freq_title.keys()))
y = np.array(list(freq_title.values()))
```

```python
plt.scatter(x,y)                            # Normal distribution graph
plt.xlabel("Movie title len")
plt.ylabel("No of movies")
plt.show()
```
