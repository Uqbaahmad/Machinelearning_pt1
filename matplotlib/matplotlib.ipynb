```python
import matplotlib.pyplot as plt
import numpy as np
```

### Ploting functions and lines

```python
a = np.linspace(0, 10, 100)
plt.plot(a, np.sin(a))
plt.show()
```

Add axis range and labels

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))

plt.xlim([2, 8])
plt.ylim([0, 0.75])

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
x = np.linspace(0, 10, 10)
plt.plot(x, np.sin(x), label = ('sin x'),linestyle = '', marker = 'o')

plt.xlabel('f(x)')
plt.ylabel('x')

plt.legend(loc = 'lower left')
plt.show()
```

### Scatter plot

```python
rng = np.random.RandomState(123)
x = rng.normal(size=100)
y = rng.normal(size= 100)

plt.scatter(x, y)
plt.show()
```

### Bar plots

```python
# input data 
means = [5, 8, 10]
stddevs = [0.2, 0.4, 0.5]
bar_labels = ['bar1', 'bar2', 'bar3']

# plot graph
x = list(range(len(bar_labels)))
plt.bar(x, means, yerr = stddevs)

plt.show()
```

### Histrogram

```python
my_list = [1, 2, 3, 1, 2, 4, 5, 4 ,6, 2]
print("List Before ", my_list)
temp_list = []

for i in my_list:
    if i not in temp_list:
        temp_list.append(i)

my_list = temp_list

print("List After removing duplicates ", my_list)
```
