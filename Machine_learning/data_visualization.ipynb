```python
# lineplot
x = np.arange(10)
y1 = x ** 2
y2 = 2*x + 3
print(x)
print(y1)
```

```python
plt.style.use("seaborn")
x_coordinates = np.array([0, 1, 2])*2
plt.bar(x_coordinates,[10, 20,15], width=0.5, label="Current Year",tick_label=['Gold','Platinum','Silver'])
plt.bar(x_coordinates+0.5, [20, 10, 12], width=0.5, label="Next Year")
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0, 40)
plt.xlim(-2, 5)
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_coordinates = np.array([0, 1, 2])*2
plt.bar(x_coordinates,[10, 20,15], width=0.5, label="Current Year",tick_label=['Gold','Platinum','Silver'], color = 'pink')
plt.bar(x_coordinates+0.5, [20, 10, 12], width=0.5, label="Next Year", color = "purple")
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

### Data Visualisation Challenge

```python
path = 'https://raw.githubusercontent.com/coding-blocks-archives/machine-learning-june-2019/master/Dataset/movie_metadata.csv'
```

```python
df = pd.read_csv(path)
```

```python
df
```

```python
df.head(10)
```

```python
df.columns
```

```python
titles = df.get('movie_title')
```

```python
print(titles)
```

```python
titles1 = list(df.get('movie_title'))
print(type(titles1))
```

```python
titles1[:5]
```

```python
titles1[0][:-2]
```

```python
freq_title = {}

for title in titles1:
    length = len(title)

    if freq_title.get(length) is None:
        freq_title[length] = 1
    else:
        freq_title[length] += 1    
```

```python
print(freq_title)
```

```python
x = np.array(list(freq_title.keys()))
y = np.array(list(freq_title.values()))
```

```python
plt.style.use('seaborn')
plt.scatter(x, y)
plt.xlabel('Length of movie title')
plt.ylabel('Freq count')
plt.title('Movie data visualisation problem')
plt.show()
```
