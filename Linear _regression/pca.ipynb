### PCA

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import numpy as np


from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.datasets  import load_diabetes
```

```python
url = "https://raw.githubusercontent.com/digipodium/Datasets/main/classfication/diabetes.csv"
df = pd.read_csv(url)
df.head()
```

```python
X = df.drop(columns=['Outcome'])
y = df.Outcome
```

```python
scaler = StandardScaler()
```

```python
df.iloc[:,0:3] = scaler.fit_transform(df.iloc[:,0:3])
```

```python
# finding covariance metrix

covariance_matrix = np.cov([df.iloc[:,0], df.iloc[:,1], df.iloc[:,2]])
print('Covariance Matrix:\n', covariance_matrix)
```

```python
eigen_values, eigen_vector  = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
features = PCA(n_components=3).fit_transform(X, y)
```

```python
px.scatter_3d(x=features[:,0], y = features[:,1], z= features[:,2], color= y.values)
```

```python
X_train,X_test, y_trian, y_test = train_test_split(features, y.values, test_size=.2, random_state=2)
clf = make_pipeline(StandardScaler(), KNeighborsClassifier())
clf.fit(X_train, y_trian)
y_pred = clf.predict(X_test)
print('Matric\n', confusion_matrix(y_test, y_pred))
print('Report\n', classification_report(y_test, y_pred))
```
