```python
vk = delivery['batsman'] =='V kohli'
new_vk = delivery[vk]
```

```python
new_vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
def run_scored(batsman_name):
    vk = delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending= False).head()
```

```python
run_scored('V Kohli')
```

```python
run_scored('G Gambhir')
```

```python
run_scored('MS Dhoni')
```
