## Scatter plots :-

### a. relplot function

```python
# scatter plot between total_bill and tip
sns.relplot(x = "total_bill", y = "tip",kind = 'scatter', data= data)
plt.show()
```

### - hue parameter

```python
sns.relplot(x = "total_bill", y= "tip", hue ="smoker", data=data)
```

### - Style parameter

```python
sns.relplot(x ="total_bill", y = "tip", hue="smoker", style='sex', data=data)
```

### - Size parameter

```python
sns.relplot(x ="total_bill", y = "tip", hue="smoker", style='sex',size='size', data=data)
```

```python
sns.relplot(x ="total_bill", y = "tip", hue="smoker", style='sex',size='size', sizes=(15, 200), data=data)
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue='smoker', data=data)
```

## Categorical estimate plots

## Categorical categorical data

### a. heatmap

```python
flights.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
x = flights.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar=False)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot= True, fmt ='d')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot= True, fmt ='d', cmap='summer')
```

### b. clustermap

```python
sns.clustermap(x)                              # clustermap - similar rows and columns are together
```

```python
sns.clustermap(x, metric='correlation', z_score=0, annot =True)                                  
```

```python
sns.clustermap(x, metric='correlation', z_score=0, annot =True, row_cluster=False)
```

```python
sns.clustermap(x, metric='correlation', z_score=0, annot =True, col_cluster=False)
```

## Displots

```python
### Displot combines 3 things in one graph:
# 1. Histogram
# 2. kernel Density Estimation (PDF)
# 3. Rugplot
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace= True)
```

### Plotting a Displot

```python
sns.distplot(titanic['age'])
```

### Plotting only the histrogram

```python
sns.distplot(titanic['age'], kde=False)
```

```python
sns.distplot(titanic['age'], kde=False, bins=10)
```

plotting only the kde

```python
sns.distplot(titanic['age'], hist=False)
```

plotting only the Rugplot

```python
sns.distplot(titanic['age'], hist= False, kde=False, rug=True)
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
```

```python

sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
