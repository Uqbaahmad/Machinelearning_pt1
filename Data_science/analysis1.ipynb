## filtering a dataframe

```python
df['Type_1'] ==  "Grass"                             # boolean filter
```

```python
df[df['Type_1'] ==  "Grass" ]
```

```python
grass_df = df[df['Type_1'] ==  "Grass" ].copy()
```

```python
sns.jointplot(x = 'Attack', y = "Defense", data=grass_df, hue = "Type_2")
```

```python
# to comparing two filter
f1 = df['Type_1'] == "Grass"
f2 = df["Type_1"] == "Fire"
df[f1 | f2]
```

```python
import plotly.express as px
```

```python
px.scatter_3d(df[f1|f2], x ='Attack', y ='Defense',z ='Speed', color='Type_1', size = 'Total',title="Pokemon Type Fire and Grass", height=600, width= 600, hover_name='Name')
```
